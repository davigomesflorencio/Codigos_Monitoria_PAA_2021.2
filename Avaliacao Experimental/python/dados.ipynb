```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lx = np.array([25,50,100,200,400,800,1600])
ly = np.array([0.00036e-07,0.002,0.01,0.13,1.18,9.65,77.69 ])

# lx=np.log2(x)
# ly=np.log2(y)

plt.scatter(lx,ly)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(lx.reshape(-1,1),ly)
print("slope=",model.coef_)

plt.scatter(lx,ly)
plt.plot(lx,model.intercept_+model.coef_*lx,'r')
plt.show()
```

```python
import statsmodels.api as sm 

lx = sm.add_constant(lx)
res = sm.OLS(ly,lx).fit()
print("interval ",res.conf_int(0.05)[1])
```
